```python
import requests
import csv
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup 
url = 'https://www.tripadvisor.com/Hotels-g60745-Boston_Massachusetts-Hotels.html'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
containers = soup.find_all('div', class_='listing collapsed')
titles = []
provider = []
price = []
url = []
review = []

for container in containers:
    titles.append(container.find('a', class_='property_title prominent ').text.replace(",", ";").strip())
    provider.append(container.find('div', class_='main-cols').find('div', {'data-sizegroup':'mini-meta-provider'}).text.replace(",", ";").strip())

    price.append(container.find('div', {'data-sizegroup':'mini-meta-price'}).text.replace(",", ";").strip())
    
    url.append(container.find('a', class_='property_title prominent ')["href"].replace(",", ";").strip())
    review.append(container.find('a', class_='review_count').text.replace(",", ";").strip())

user1_data = {'title': titles, 
             'provider':provider,
             'price':price, 
             'url':url, 
             'review':review}
df1 = pd.DataFrame(user1_data, columns = ['title', 'provider', 'price', 'url','review'])
df1.head(df1.size)
```
